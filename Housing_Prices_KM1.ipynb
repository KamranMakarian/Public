# Feature Engineering

# Train/Test split and scaling

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

# Model creation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),
          batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.head(2)
```

```python
losses.plot()
```

# Evaluation on Test Data

https://scikit-learn.org/stable/modules/model_evaluation.html#regression-metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score,r2_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test,predictions) # this is less than 20% of the mean price. Not too bad! MAB does not punish large errors.
```

```python
mean_squared_error(y_test,predictions) # this is same as mse loss function for test data. MSA is the unit of price to the power of 2.
```

```python
np.sqrt(mean_squared_error(y_test,predictions)) # Unit of price.
```

```python
explained_variance_score(y_test,predictions)
```

```python
r2_score(y_test,predictions) # notice that it is same as the previous line.
```

```python
df['price'].mean()
```

```python
df['price'].median()
```

```python
training_score = model.evaluate(X_train,y_train,verbose=0) # this is the loss value from mse loss function. 
test_score = model.evaluate(X_test,y_test,verbose=0) # this is same as mse loss function for test data.
```

```python
print(training_score)
print(test_score)
```
